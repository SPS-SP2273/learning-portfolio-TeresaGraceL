# 4 A complete OO example

```python
#simple example with a nicer, tweaked version of the previous plot

fig, ax = plt.subplots(nrows=2, ncols=1,
                       figsize=(5, 5),
                       sharex=True)


ax[0].plot(x, cos_x, label=r'$\cos(x)$')
ax[0].fill_between(x, 0, cos_x, alpha=.25)
ax[1].plot(x, sin_x, label=r'$\sin(x)$')
ax[1].fill_between(x, 0, sin_x, alpha=.25)

for a in ax:
    a.legend()
    a.grid(alpha=.25)
    a.set_ylabel('$y$', rotation=0)


ax[1].set_xlabel('$x$')

fig.suptitle(r'$\sin(x)$ and $\cos(x)$')

fig.tight_layout()
```

**How to plot using OO**
- Set up the axes and figure
    - Get two axes ready to plot in two rows in one column.
    - Change the size of the figure by specifying a figure size (figsize).
    - Ask that the plots share the $x$-axis using sharex.
- Since ax is a NumPy array with two axes, we can index them using 0 and 1. Then we just plot to that axis.
- Use fill_between() to fill the plots.
    - again just indexed the NumPy array to access the axes.
- Draw the legends
    - As in the previous example, we can do this one axis at a time. However, a more sensible way to do this is with a for loop that iterates through the items in ax.
    - Letâ€™s also add a grid to each plot and set the label.
- We are sharing the $y$-axis. So, we only need to label the lowest plot. This has the index 1.
- Letâ€™s add a super title to the figure (notplot).
- Finally, letâ€™s ask Matplotlib to make any necessary adjustments to the layout to make our plot look nice by calling tight_layout(). tight_layout() optimises the space in the plot to make it neater and better looking.

```python
#further modification of the above example

fig, ax = plt.subplots(nrows=2, ncols=1,
                       figsize=(5, 5),
                       sharex=True)

ax[0].plot(x, cos_x, label=r'$\cos(x)$') 
    #r in front for it to ignore the slash instead of thinking its escape sequence
ax[0].fill_between(x, 1, cos_x, alpha=.25, color ='g') #x values and limits you want filled

ax[1].plot(x, sin_x, label=r'$\sin(x)$')
ax[1].fill_between(x, 0, sin_x, alpha=.25)

#---------------- grid - use a for loop instead
#ax[0].grid(alpha.25)
#ax[1].grid(alpha.25)

for a in ax.flatten():
    a.legend()
    a.grid(alpha=.25)
    a.set_ylabel('$y$', rotation=0)

ax[1].set_xlabel('$x$')
ax[1].set_title('title')

fig.suptitle(r'$\sin(x)$ and $\cos(x)$') #title belongs to figure and not axis
fig.tight_layout() #very useful, cleanup crew because otherwise there is a big space

plt.savefig('plotting_good_simpleplot.png', dpi=150)
```

## 5.1 Sharing axes - some examples

- make the plots more compact by sharing the and axes using sharex and sharey.

```python
#what happens if not specified how to share
fig, ax = plt.subplots(
    nrows=2, ncols=2,
    figsize=(5, 5)
)
```

Without specifying how to share, Matplotlib auto-scaled both axes 
- plots on the left go from 0 to 4, and those on the right go from 0 to 5
- lengths that were used for the lines

```python
#not easily readable
fig, ax = plt.subplots(nrows=2, ncols=2,
                       figsize=(5, 5),
                       )
top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.hlines(y=1, xmin=0, xmax=5)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=5)
bottom_right.hlines(y=4, xmin=0, xmax=5)
```

specifying how to share axes
- True: **all** axes use same range
- col: use the same range for all columns
- row: use the same range for all rows

```python
#all the plots with same range of the x axis
fig, ax = plt.subplots(
    nrows=2, ncols=2,
    figsize=(5, 5),
    sharex=True, sharey='row'
)
top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.hlines(y=1, xmin=0, xmax=5)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=5)
bottom_right.hlines(y=4, xmin=0, xmax=5)
```

- sharex='col' is more suited for the data we are plotting

```python
fig, ax = plt.subplots(
    nrows=2, ncols=2,
    figsize=(5, 5),
    sharex='col', sharey='row'
)

top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.hlines(y=1, xmin=0, xmax=5)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=5)
bottom_right.hlines(y=4, xmin=0, xmax=5)
```

## 6.2 Scatter plots

Scatter plots created by putting a marker at an point you specify -- simple yet powerful.

- Using the same data as above, need some values for x
- use range() along with len() to generate a list [0,1,2...] appropriate to the dataset.

```python
plt.style.use("seaborn-v0_8-darkgrid")

data = get_plot_data()

for plot_label, numbers in data.items():
    x = range(len(numbers))
    y = numbers
    plt.scatter(x, y, label=plot_label, alpha=.5)

plt.xlabel('Position in the list')
plt.ylabel('Mean of Basic Montly Eraning (S$)')
plt.legend()
```

## 6.4 Pie charts ('not very quantitative')

using the same dummy data as above:

```python
student_numbers = {'Life Sciences': 14,
                   'Physics': 12,
                   'Chemistry': 8,
                   'Comp. Biology': 1}
majors = list(student_numbers.keys())
numbers = list(student_numbers.values())

plt.style.use('fivethirtyeight')
plt.pie(numbers, 
        labels=majors,
        autopct='%1.3f%%',   # How to format the percentages, number of decimal places
        startangle=-90       #rotate the pie chart in any way you like         
        )
plt.title('Percentage of each major')
```
