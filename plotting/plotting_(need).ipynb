## 2.1 Adding mathematical functions

- note: if we use the sin from another package, this will not work!
- must use the numpy sin and not the math sin
- the sin from other packages cannot handle a whole list of numbers

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2                  # The math stuff
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )
plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend()                                 
```

this is a horrible plot - the blue line is not clear.

## We need another axis!

- add another y axis that shares the same x axis using the command twinx()

When having two axis need to be careful, a better option might be to have two plots

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1, you must run the legend before running twinx

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2
```

- this sin curve does not look curvy
- need to have better resolution -- get Numpy to create a list of more numbers using arange OR use linspace

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x = np.arange(0, 6, .25)

x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc=(0.25,0.5), frameon=False)                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2
plt.legend(loc=(0.25,0.45), frameon=False)                    # to make the legends in a similar place so it looks more seamless
```

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x = np.linspace(0, 6, 100)

x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2
```

## 4.1 Plotting data from files

Hockeystick plot of average global land temperature against the year
- hlines function needs a y value and starting and ending values for x

```python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey') #add a horizontal line at zero value to highlight trend shown
plt.plot(date, anomaly, alpha=.5)
#plt.ylim([-8, 8])
plt.grid(alpha=.25)
```

## 4.2 Styles

Matplotlib offers some standard style templates, to get a decent looking plot without customising it.

plt.style.use('') must be included right at the top. This is the [list of Maplotlib styles](https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html)

```python
plt.style.use('fivethirtyeight')
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
```
