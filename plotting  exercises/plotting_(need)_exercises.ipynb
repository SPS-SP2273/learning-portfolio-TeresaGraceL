# Exercise 1 Fill in the blanks to produce the plot

https://matplotlib.org/stable/api/markers_api.html

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr= x_err, yerr= y_err, #error values according to the list
             color='darkred', linestyle='--',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label = 'Y values with errors') #label for the legend
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right') #position of the legend
plt.show()
```

# Exercise 2 A simple Mathematical plot

```python
import numpy as np
```

Use NumPy to generate data suitable for plotting the graph of the function: $$y=e(^(-x/10)^)sinx$$
for values of in the range 0 to 50.

Customise your plot so that:

- The x and y axes have labels of fontsize of 15
- The x and y axes labels are in the color, darkorange.
- There is a grid with an opacity of 25%.
Indicate the equation of the function in the title with a font size of 20.
    - Hint: Look at [Writing mathematical expressions in Matplotlib.](https://matplotlib.org/stable/tutorials/text/mathtext.html)
- Change the limit (using ylim) to [-1,1]
- In the end, your plot should look like the one shown alongside.

```python
x = np.linspace(0, 50, 500)  # #Generate an array of 500 x values so it will be smooth
y = np.exp(-x / 10) * np.sin(x) #Calculate y values using the formula

plt.plot(x, y)
plt.xlabel('x', fontsize=15, color='darkorange')
plt.ylabel('y', fontsize=15, color='darkorange')
plt.title('$y = e^{-x/10}sin(x)$', fontsize = 20 ) # title and make the text math 
plt.grid(True, alpha=0.25) #grid and opacity
plt.ylim([-1, 1]) #limit
plt.show()
```
