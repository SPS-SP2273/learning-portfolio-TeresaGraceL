```python
# Dictionary for month conversion
month_conversion = {
    'jan': '01', 'feb': '02', 'mar': '03',
    'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09',
    'oct': '10', 'nov': '11', 'dec': '12'
}

# Example filename
filename = 'oct-08_singapore.txt'

# Split the filename into components
parts = filename.split('-')
month_part = parts[0]  # Get month part
day_and_city = parts[1].split('_')  # Split the day and city part
day_part = day_and_city[0]  # Get the day part
city_part = day_and_city[1].split('.')[0]  # Get the city part without file extension

# Convert month part and reassemble the filename
converted_month = month_conversion[month_part]
new_filename = f"{converted_month}-{day_part}-{city_part}.txt"

print(new_filename)
```

# Task 2

Incorporate your previous code into a function named rename_my_file(old_file_name) that accepts the old filename as the argument and returns the new file name.

```python
month_conversion = { #dictionary, leaving it out of the function
    'jan': '01', 'feb': '02', 'mar': '03',
    'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09',
    'oct': '10', 'nov': '11', 'dec': '12'}

file_name = 'oct-08_singapore.txt'

def rename_my_file(old_file_name):
    parts = old_file_name.split('-')
    month_part = parts[0] 
    day_and_city = parts[1].split('_') 
    day_part = day_and_city[0] 
    city_part = day_and_city[1].split('.')[0]
    converted_month = month_conversion[month_part]
    new_filename = f"{converted_month}-{day_part}-{city_part}.txt"
    
    return new_filename
```
